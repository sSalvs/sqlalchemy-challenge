# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model

Base = automap_base()

# reflect the tables

Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found

Base.classes.keys()
```

```python
# Save references to each table

Measurement = Base.classes.measurement

Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB

session = Session(engine)
```

```python
#Inspect the engine

inspector = inspect(engine)
```

```python
#look at the tables and columns 

inspector.get_table_names()
```

```python
inspector.get_columns('measurement')
```

```python
inspector.get_columns('station')
```

# Precipitation Analysis

```python
#Find the last date 

last_date = session.query(func.max(Measurement.date)).first()
last_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database

year_ago = dt.date(2017, 8, 23) - dt.timedelta(days=365)
year_ago
```

```python
# Perform a query to retrieve the date and precipitation scores

twelve_months_precip = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= year_ago).all()
twelve_months_precip
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column

precip_df = pd.DataFrame(twelve_months_precip, columns=["date", "precip"])

# set the index to date for the graph 

precip_df = precip_df.set_index("date")

# Sort the dataframe by date

precip_sorted = precip_df.sort_values("date")
precip_sorted
```

```python
# Use Pandas Plotting with Matplotlib to plot the data

precip_sorted.plot()
plt.ylabel("Inches")
plt.xticks(rotation='vertical')
plt.savefig('Outputs/date_vs_precip')
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data

precip_df.describe()
```

# Station Analysis

```python
# Design a query to show how many stations are available in this dataset?

station_result = session.query(func.count(Station.station)).all()

print(station_result)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

active_stations = session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
active_stations

```

```python
# Which station has the highest number of observations?

print(active_stations[0])
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

sel = [Measurement.station, 
       func.min(Measurement.tobs), 
       func.max(Measurement.tobs), 
       func.avg(Measurement.tobs)]

recorded_temps = session.query(*sel).filter(Measurement.station == "USC00519281").all()
# print the results

for temp in recorded_temps:
    print(f"Station ID {temp[0]}")
    print(f"the min is {temp[1]}")
    print(f"the max is {temp[2]}")
    print(f"the average is {temp[3]}")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram


twelve_months_obvs = session.query(Measurement.date, Measurement.tobs).filter((Measurement.date >= year_ago) & (Measurement.station == "USC00519281")).all()
# twelve_months_obvs

# Create a dataframe from the results 
temp_obvs_df = pd.DataFrame(twelve_months_obvs, columns =["date", "tobs"])
temp_obvs_df

# Plot a histogram 
ax = temp_obvs_df.plot.hist(bins=12)
ax.set_xlabel("Temparature")
plt.savefig('Outputs/temp_hist')
plt.show()
```
